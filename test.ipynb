```python
import numpy as np
import argparse
import yaml
import sys
import matplotlib.pyplot as plt

import tool as tl
import mhpspike
import fit
with open("./_out/sample/best_params.yaml") as file:
        dic = yaml.safe_load(file)

with open("./_out/sample/params.yaml") as file:
        hparams = yaml.safe_load(file)
```

```python
T = hparams['T']
params=np.zeros(9)
params[0] = float(dic['N_max'])
params[1] = float(dic['betaN'])
params[2] = float(dic['slope'])
params[3] = float(dic['nc'])
params[4] = float(dic['Sc'])
params[5] = float(dic['bgnoise'])
params[6] = float(dic['Pp'])
params[7] = float(dic['Pa'])
params[8] = float(dic['Ps'])
(idx,b,u)=mhpspike.spikeM(params, T)

data = tl.loadseq("./_data/sample/sequence.dat").T
dat = data.flatten()
dat = dat[:T]
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('t')  # x軸ラベル
ax.set_ylabel('y')  # y軸ラベル
# ax.set_title(r'$\sin(x)$ and $\cos(x)$') # グラフタイトル
ax.plot(b,label = 'b')
#ax.plot(u,label = 'u')
ax.plot(dat,label = 'dat')
plt.yscale("log")
plt.xscale("log")
plt.legend(loc = 'upper right')
ax.set_xlim([10,96])
plt.show()
```
